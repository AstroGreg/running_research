## Lets start with loading in the libraries

```python
import pandas as pd
# import librabry to display data in a tablex
import matplotlib.pyplot as plt
```

```python
# Read the data from file using data/day_approach.csv
df = pd.read_csv('data/week_approach.csv')
df
```

```python
# explore the columns
columns = df.columns
# import regex library
import re

# with .2  means 3 weeks before injury 
two_weeks_before_evaluation = list(filter(lambda x: re.search(r'\.2$', x), columns))
one_week_before_evaluation = list(filter(lambda x: re.search(r'\.1$', x), columns))
week_of_evaluation = list(filter(lambda x: x not in two_weeks_before_evaluation and x not in one_week_before_evaluation, columns))

# create a new dataframe with the columns
df_two_weeks_before_evaluation = df[two_weeks_before_evaluation]
df_one_week_before_evaluation = df[one_week_before_evaluation]
df_week_of_evaluation = df[week_of_evaluation]
df_week_of_evaluation
```

```python
# find injured runners 

injured_runners = df_week_of_evaluation[df_week_of_evaluation['injury'] == 1]

```

```python
# compare every column with the injury column
# find the columns with the highest correlation with the injury column0
from scipy.stats import pearsonr


# go over all the columns and find the p-value of the correlation make a dataframe of it

df_corr = pd.DataFrame(columns=['column', 'correlation', 'p-value'])
for column in df_week_of_evaluation.columns:
    if column != 'injury':
        correlation, p_value = pearsonr(df_week_of_evaluation[column], df_week_of_evaluation['injury'])
        if p_value < 0.05:
           df_corr = pd.concat([df_corr, pd.DataFrame([[column, correlation, p_value]], columns=['column', 'correlation', 'p-value'])])
df_corr
```
